## **NDITER**



1.   nditer()-used for n for loops.(in the above iterations in 3-D array we use 3 for loop iteration.if we want n iteration we use this)

## Range,Variance,StandardDeviation,IQR

```python
import numpy as np
import scipy.stats as stats
data=[12,15,14,10,18,20,22,25,14,17]

#Range
range_value=max(data)-min(data)
print(f"Range : {range_value}")
print()

#variance
variance_value=np.var(data,ddof=0)
sample_variance_value=np.var(data,ddof=1)
print(f"Population Variance: {variance_value}")
print(f"Sample Variance: {sample_variance_value}")
print()

#standard deviation
std_dev_value=np.std(data,ddof=0)
sample_std_dev_value = np.std(data, ddof=1)
print(f"Population Standard Deviation: {std_dev_value}")
print(f"Sample Standard Deviation: {sample_std_dev_value}")
print()

#IQR
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
iqr_value = Q3 - Q1
print(f"Interquartile Range (IQR): {iqr_value}")
```

# NUMPY UFUNCS



1.   universal functions they are numpy functions that operates on the **ndarray** object
2.   ufuncs are used to implement vectorization



## vectorization

converting iterative stmt into a **vector based operations**

1. without ufunc,zip() - combines elements from two or more sequences **pair-wise**.

```python
#add the elements of 2 lists
x=[1,2,3,4]
y=[4,5,6,7]
z=[]
for i ,j in zip(x,y):
  z.append(i+j)
print(z)
```

2. with ufunc,use **add()** fun

```python
import numpy as np
x=[1,2,3,4]
y=[5,6,7,8]
z=np.add(x,y)
print(z)
```

```python
import numpy as np
x=np.array([1,2,3,4])
y=np.array([5,6,7,8])
print(np.add(x,y))
```

3. your own ufunc

```python
#create your own ufunc for addition
import numpy as np
def myadd(x,y):
  return x+y
myadd=np.frompyfunc(myadd,2,1)
print(myadd([1,2,3,4],[5,6,7,8]))
```

## union

**union1d()**

```python
#find union of the following 2 set arrays
import numpy as np
a1=np.array([1,2,3,4])
a2=np.array([3,4,5,6])
newa=np.union1d(a1,a2)
print(newa)
```

## intersection

**intersect1d()**

```python
#find intersection of the following 2 set arrays
import numpy as np
a1=np.array([1,2,3,4])
a2=np.array([3,4,5,6])
newa=np.intersect1d(a1,a2,assume_unique=True)
print(newa)
```

## finding difference

**setdiff1d()**-only the values in the first set that is **not** present second set

```python
#find the diff of the set1 from set2
import numpy as np
s1=np.array([1,2,3,4])
s2=np.array([3,4,5,6])
news=np.setdiff1d(s1,s2,assume_unique=True)
print(news)
```

## finding symmetric difference

**setxor1d()**-only the values that are **not** present in **both**

```python
#find the symmetric difference of the set1 and set2
import numpy as np
s1=np.array([1,2,3,4])
s2=np.array([3,4,5,6])
news=np.setxor1d(s1,s2,assume_unique=True)
print(news)
```

## 41:Filter with Multiple Conditions: Filter the array a = np.array([10, 15, 20, 25, 30]) to get values that are both greater than 15 and less than 30.

```python
import numpy as np
a = np.array([10, 15, 20, 25, 30])
result = a[(a > 15) & (a < 30)]
print(result)
```

## 42: Filtering a 2D Array: Filter the 2D array a = np.array([[1, 2], [3, 4], [5, 6]]) to only include rows where the first column is greater than 2

```python
import numpy as np

a = np.array([[1, 2],
              [3, 4],
              [5, 6]])

filtered_a = a[a[:, 0] > 2]
print(filtered_a)
```
